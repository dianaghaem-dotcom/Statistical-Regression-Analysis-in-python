# Generate missing values

```python
import numpy as np

df_train.head()

np.random.seed(42)
indexes_day=np.random.choice(df_train.index, size=int(0.05*len(df_train)), replace=False)

indexes_temperature=np.random.choice(df_train.index, size=int(0.05*len(df_train)), replace=False)


```

```python
df_train.loc[indexes_day, "Day of Week"] = np.nan
df_train.loc[indexes_temperature, "Average Temperature"] = np.nan

```

```python
# محاسبه میانگین و پر کردن مقادیر خالی دمای هوا
temp_mean = df_train["Average Temperature"].mean()
df_train["Average Temperature"].fillna(temp_mean, inplace=True)

# پیدا کردن مد و پر کردن مقادیر خالی روز هفته
day_mode = df_train["Day of Week"].mode()[0]
df_train["Day of Week"].fillna(day_mode, inplace=True)

# چک کردن نهایی
df_train.isnull().sum()
```
